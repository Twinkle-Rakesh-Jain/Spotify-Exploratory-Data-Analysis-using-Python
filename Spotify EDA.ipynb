```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('/Users/twinklejain/Desktop/Data Analysis Project/Spotify Data Analysis/tracks.csv')
df_tracks.head()
```

```python
#checking for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# Finding top 10 least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
#Finding out the descriptive statistics - count, min, max, std etc.
df_tracks.describe().transpose()
```

```python
#Finding out top 10 popular songs with popularity greater than 90. The query method is used to filter the dataframe based on a condition. The parameter 'inplace =False' specifies that the filtering operation will not modify the original dataframe and will return the new dataframe
most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending= False)
most_popular[:10]
```

```python
#Changing index to release date in a datetime format
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#Checking Artist name at a particular index
df_tracks[["artists"]].iloc[18]
```

```python
#Converting duration in millisecs to secs
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace= True, axis = 1) #axis = 1 refers to columns which means that the operation should be performed across columns
```

```python
df_tracks.duration.head()
```

```python
#First Visualisation: Correltaion matrix between columns except key, mode and explicit using Pearson coefficient which measures linear relationship between two variables. It then creates a heatmap to visualise the correlation between the remaining variables.
corr_df = df_tracks.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Correlation plot examines relationships between multiple pairs of variables simulataneously. It can be seen that there is a high positive correlation between Loudness and Energy - 0.8
```

```python
#Creating sample data - 0.4% of the total data
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
#Creating regression plot between Energy and Loudness Columns using sample data from above. Regression plot focuses only on two variables
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y= "loudness", x="energy", color='c').set(title= "Loudness vs Energy Correlation")
```

```python
#The fact that points are clustered around the line indicates that the two variables are closely related. The upward slope of the line means that as the value of the energy increases, the value of loudness also tends to increase. So loudness and energy have a strong positive linear relationship
```

```python
#Creating regression plot between Popularity and Acousticness Columns using sample data from above. Regression plot focuses only on two variables
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y= "popularity", x="acousticness", color='b').set(title= "Popularity vs Acousticness Correlation")
```

```python
#Since the points are widely distributed and not tightly clustered around the regression line, it suggests that the relationshsip between popularity and acousticness is weaker. The independent variable (acousticness) does not strongly predict the dependent variable (popularity). The downward slope indicates that there is a negative relationship between these two variables. As the value of acousticness increases, the value of popularity tends to decrease
```

```python
#Creating years column from release date column
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
#Creating a distribution plot to visualise the total number of songs in each year since 1922
#pip install --user seaborn==0.11.0
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
# We will visualise duration of songs over the years using barplot
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax =plt.subplots(figsize = fig_dims)
fig= sns.barplot(x=years, y=total_dr, ax=ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90) #This rotates the x-axis labels by 90 degrees so they are displayed vertically. This is often done when there are many labels that would overlap if displayed horizontally
```

```python
#We can observe that in the 1920s, the average duration of tracks was shorter. Over time, there was a gradual increase in duration, followed by a period of relative consistency up until 2021, with only minor fluctuations along the way.
```

```python
#We will visualise the song duration over the years using line plot now
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax =plt.subplots(figsize = fig_dims)
fig= sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
#"We can observe that songs from the 1900s to the 1960s generally had shorter durations, likely because most songs during this period were primarily vocal rather than instrumental. After the 1960s, song durations began to increase, peaking around 2010. However, post-2010, there's a noticeable decline in song durations, which may be attributed to the decreasing average attention span of listeners.
```
